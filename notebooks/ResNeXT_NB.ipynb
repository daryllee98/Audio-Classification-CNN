```python
import resnext
```

```python
#help(resnext.ResNext)
```

```python
image_shape = (112, 112, 3)
depth = 29
cardinality = 8
width = 64
weight_decay = 5e-4
classes = 10


model = resnext.ResNext(image_shape, depth, cardinality, width, weight_decay)
```

```python
#model.layers
```

```python
from keras.utils.vis_utils import plot_model
from IPython.display import Image

from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
```

```python
plot_model(model, to_file='model_plot_resnext.png', show_shapes=True, show_layer_names=True)
Image(filename='model_plot_resnext.png')
```

```python
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```
